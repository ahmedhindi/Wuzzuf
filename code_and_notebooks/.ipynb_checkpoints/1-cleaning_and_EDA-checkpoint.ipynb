```python
#reading data
data = pd.read_csv('data.csv', na_values='Select')
```

#### Parsing datetime

```python
datetime = pd.to_datetime(data.post_date)
temp = pd.DatetimeIndex(datetime)
data['date'] = temp.date
data['time'] = temp.time
data.set_index(pd.to_datetime(data.post_date),inplace=True)
data.drop('post_date',axis=1, inplace=True)
```

#### Fuzzy matching for the city_name 

```python
matchs_list = list(data.city_name.value_counts().head(5).index) #most common values
data['city_name'] = data.city_name.apply(fuzzy_match, matchs_list=matchs_list)
```

####  Parsing the description, job_requirements coloums and removing the stop words

```python
data['description'] = data.description.apply(parse_html)
data['job_requirements'] = data.job_requirements.apply(parse_html)
```

```python
num_posts = data.groupby('date').count().id
```

### While playing with the data i wondered does the number of filled data  affect the number of views?
##### so let's see that. 

```python
#sum the number of unfilled celles for every job post 
alist = ['job_category_1', 'job_category_2', 'job_category_3', 'job_industry_3','job_industry_1', 'job_industry_2']
data['unfilled_info'] = data[alist].apply(pd.isnull).apply(sum,axis=1)
```

#### The number of views for every  the number of the unfilled data

```python
f = ["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]
data.groupby('unfilled_info')['views'].mean()[:5].plot(kind='bar',  
                                                       title='the number of views based on the number of the unfilled data',
                                                       color=f,figsize=(15,5));
```

#### It doesn't seem to have any significant effect however let's check the frequencies of every case of the number of unfilled cells.

```python
ax = data.groupby('unfilled_info').agg('count')['count'].plot(kind='bar', figsize=(15,5), color=f)
ax.set_xlim(-.5,4.5);
```

well, the data is highly imbalanced but it's enough to be Representative. 
and the conclusion is that it's according to this data (and just this data) the number of filled cells from the columns ('job_category_(1,2,3)', 'job_industry_(1,2,3)') does not appear to be significant determinant of the number of views.
